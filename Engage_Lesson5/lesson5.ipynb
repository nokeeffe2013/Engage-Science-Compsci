Space Mission: You have been tasked by the ESA (European Space Agency) to calculate key information for an upcoming space mission.
You should be assigned a planet and work in a group to successfully generate the calculations for this mission.

The first thing to discover is the key numbers for this mission, such as gravity on Earth, in space and on the destination planet.
weight (N) = mass (kg) x  gravitational field strength (N/kg).

```python
# Constants
gravity_earth = 9.81  # Gravitational field strength on Earth (N/kg)
gravity_space = 0     # Microgravity approximation (N/kg)

# Gravitational field strengths for planets
gravity_planets = {
    'Mars': 0,  # N/kg
    'Jupiter': 0,  # N/kg
    'Saturn': 0   # N/kg
}

# Function to calculate weight (force of gravity)
def calculate_weight(mass, gravity):
    return mass * gravity
```

Next we need to store information about the spacecraft, such as the destination planet and how heavy it is.

```python
# Inputs for a sample mission
planet_name = 'Mars'  # Example planet
spacecraft_mass = 1000  # kg
```

Whats the weight of the craft in each of the three locations?

```python
# Weight Calculations
weight_on_earth = calculate_weight(spacecraft_mass, gravity_earth)
weight_on_planet = 0 # Hint use gravity_planets[planet_name]
weight_in_space = 0
```

The code below creates a chart showing the weight of the spacecraft on all three locations? Does it look correct?

```python
# Gravitational Force Comparison Plot
plt.subplot(1, 2, 2)
locations = ['Earth', planet_name, 'Space']
weights = [weight_on_earth, weight_on_planet, weight_in_space]
plt.bar(locations, weights, color=['blue', 'red', 'green'])
plt.title('Weight Comparison')
plt.xlabel('Location')
plt.ylabel('Weight (N)')

plt.tight_layout()
plt.show()
```

Section 2: Fuel Consumption
Change the numbers to solve the question in the worksheet - how much fuel would it take to get to your destination planet?
Time(s) = distance / speed , assume speed is 17,500 mph

```python

# Fuel Consumption Simulation
initial_fuel = 500  # kg
burn_rate = 0.5     # kg/s
speed = 17500
distance = 1

time = np.linspace(0, distance/speed, 500)  # 500 initial fuel. Hint increase this.
fuel_remaining = initial_fuel - burn_rate * time

# Visualization
plt.figure(figsize=(10, 5))

# Fuel Over Time Plot
plt.subplot(1, 2, 1)
plt.plot(time, fuel_remaining)
plt.title('Fuel Consumption Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Fuel Remaining (kg)')
```
