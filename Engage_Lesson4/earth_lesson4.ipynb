First we should look at what the data contains, and it's quality. The file is called temperatureChange.txt

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Open the file
def read_file(filename) :
    with open(filename, "r", encoding="utf-8") as f:
        # Read the data into a DataFrame
        df = pd.read_csv(f, delim_whitespace=True)
        return df

# First we need to read the data from the file
df = read_file("temperatureChange.txt")
```

```python
# Plotting the data
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['No_Smoothing'], label='No Smoothing', marker='o')
plt.plot(df['Year'], df['Lowess(5)'], label='Lowess(5)', marker='x')
plt.xlabel('Year')
plt.ylabel('Temperature Change (C)')
plt.title('Land-Ocean Temperature Index (1880-2024)')
plt.legend()
plt.grid(True)
plt.show()
```
